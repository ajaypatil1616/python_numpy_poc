# POC Image processing


```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open("ajay.png")
```

```python
width, height = image.size
print(f'size of image : {image.size}')
print(f'width of image : {image.width}')
print(f'height of image : {image.height}')
```

```python
image_array = np.asarray(img)
plt.imshow(image_array)
```

```python
# print(image_array)
```

```python
sliced_array = image_array[32:600, 620:1299]
```

```python
plt.imshow(sliced_array)
```

```python
save_image = Image.fromarray(sliced_array)
save_image.save("sliced.png", format = "png")
```

```python
rotate_image = np.rot90(sliced_array)
rotated_final = Image.fromarray(rotate_image)
plt.imshow(rotate_image)
rotated_final.save("rotated.png", format= "png")
```

```python
flip_image = np.flip(sliced_array, axis=0)
final_flipped = Image.fromarray(flip_image)
plt.imshow(final_flipped)
final_flipped.save("fliped_row.png", format ="png")
```

```python
flip_image = np.flip(sliced_array, axis=1)
final_flip = Image.fromarray(flip_image)
plt.imshow(final_flip)
final_flip.save("flipped_column.png", format= "png" )
```
